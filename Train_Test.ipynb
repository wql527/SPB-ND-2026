```python
from ultralytics import YOLO
from ultralytics.engine.curriculum_callback import CurriculumCallback

model = YOLO("yolov8l_CA.yaml")

model.add_callback(
    "on_train_epoch_start",
    CurriculumCallback()
)

model.train(
    data="/content/dataset/06_data.yaml",
    epochs=100,
    imgsz=640,

    close_mosaic=10   # 最后 10 个 epoch 强制无增强,不改变原始策略
)
```
